```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_minute.html'
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr',limit=2)[0].findAll('th')]
headers = headers[1:]
headers
```

```python
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
                for i in range(len(rows))]
```

```python
per_36 = pd.DataFrame(player_stats,columns = headers)
per_36.head()
```

```python
per_36.isna().sum()
per_36.fillna(0)
```

```python
per_36.dtypes
```

```python
columns1 = ['Age','G','GS']
columns2 = ['MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS']
```

```python
for col in columns1:
    per_36[col] = per_36[col].astype(float)
```

```python
per_36.dtypes
per_36.loc[per_36['3P%'] == '']
per_36.replace('',0)
```

```python
import numpy as np
per_36 = per_36.replace(r'^\s*$', np.NaN, regex=True)
```

```python
per_36.isna().sum()
per_36.fillna(0)
```

```python
for col in columns2:
    per_36[col] = per_36[col].astype(float)
```

```python
from sqlalchemy import create_engine
```

```python
per_36.to_sql('per36',engine,if_exists='append')
```
